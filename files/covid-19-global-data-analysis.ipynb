```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
"""Il dataset contiene informazioni mondiali relative al COVID-19,
tra cui casi totali, nuovi casi, vaccinazioni, ricoveri e altri indicatori sanitari.
Sono presenti valori nulli in diverse colonne, dovuti alla mancanza di rilevazioni
per alcune nazioni o periodi temporali.
La colonna date è stata convertita in formato datetime per facilitare le analisi temporali."""
```

```python
df_continenti = df[df["continent"].notna()]
```

```python
casi_continente = df_continenti.groupby("continent")["new_cases"].sum()
```

```python
casi_continente
```

```python
#ordiniamo i continenti dal più alto al più basso
casi_continente = casi_continente.sort_values(ascending=False)

casi_continente
```

```python
tot_mondiali= casi_continente.sum()
```

```python
tot_mondiali
```

```python
percentuale_continente = (casi_continente / tot_mondiali) * 100
```

```python
percentuale_continente
```

```python
percentuale_continente = percentuale_continente.round(2)
```

```python
percentuale_continente
```

```python
"""Per l’analisi sono stati selezionati i dati relativi all’Italia nell’anno 2022.
Sono state eliminate le righe in cui new_cases presenta valori nulli, poiché rappresentano giorni senza misurazioni.

La colonna total_cases rappresenta il totale cumulativo dei casi dall’inizio della pandemia,
mentre new_cases rappresenta i nuovi casi registrati in una specifica data.
Per questo motivo total_cases mostra un andamento progressivo nel tempo,
mentre new_cases mostra variazioni e picchi legati alle singole rilevazioni."""
```

```python
df[["location", "date", "icu_patients"]].sample(5)
```

```python
paesi = ["Italy", "Germany", "France"]
```

```python
icup_df = df[
    (df["location"].isin(paesi)) &
    (df["date"] >= "2022-05-01") &
    (df["date"] <= "2023-04-30")
]
```

```python
icup_df["icu_patients"].isna().sum()
```

```python
icup_df = icup_df[icup_df["icu_patients"].notna()]
```

```python
icup_df[["date","location","icu_patients"]]
```

```python
icup_df.boxplot(
    column="icu_patients",
    by="location",
    figsize=[7,10]
)
```

```python
"""Dal boxplot emerge che Germania e Francia hanno avuto un numero di pazienti in terapia intensiva significativamente più alto rispetto all’Italia nel periodo considerato.
La Germania mostra anche una maggiore variabilità dei valori ICU, mentre l’Italia presenta valori mediamente più bassi e più concentrati."""
```

```python
df[["location", "date", "hosp_patients"]].head()
```

```python
paesi = ["Italy", "France", "Spain", "Germany"]
```

```python
hosp_2021 = df[
    (df["location"].isin(paesi)) &
    (df["date"].dt.year == 2021)
]
```

```python
hosp_2021["hosp_patients"].isna().sum()
```

```python
somma_hosp = hosp_2021.groupby("location")["hosp_patients"].sum()

somma_hosp
```

```python
somma_hosp.plot(
    kind="bar",
    figsize=(9,6),
    title="Somma dei pazienti ospedalizzati nel 2021"
)
```

```python
"""Dal grafico emerge che la Francia presenta il valore più alto di pazienti ospitalizzati nel 2021, seguita dall’Italia e dalla Spagna.
La Germania non mostra valori nel grafico, probabilmente a causa della presenza di dati nulli o mancanti nella colonna hosp_patients, non gestibili tramite sostituzione."""
```
